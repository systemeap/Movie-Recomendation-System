```python
# Haciendo un listado del diccionario que contiene, vemos que tiene muchos datos NaN, 40972 en total
# me creo una funcion, para crear diccionario vacios
#Definir un diccionario por defecto para reemplazar los valores NaN
default_dict = {'id': np.nan,'name': np.nan,'poster_path': np.nan,'backdrop_path': np.nan}
def replace_nan_dict(dic1):
    if pd.isna(dic1):
        return default_dict
    return dic1

def safe_literal(dic1):
    try:
        return ast.literal_eval(dic1)
    except (ValueError, SyntaxError):
        return dic1

# Convertir las cadenas en literales y reemplazamos NaN con el diccionario predeterminado
df_movies['belongs_to_collection'] = df_movies['belongs_to_collection'].apply(safe_literal).apply(replace_nan_dict)
print(df_movies['belongs_to_collection'])
```

```python
# Separar los componentes del diccionario en columnas nuevas
df_coll_new = df_movies['belongs_to_collection'].apply(pd.Series)                       
df_coll_new.columns
```

```python
#Eliminamos las columnas que no necesitamos
df_coll2=df_coll_new.drop(columns=[0,'id',  'poster_path', 'backdrop_path'],inplace=False)
df_coll_fin = df_coll2.rename(columns = {'name' : 'name_coll'})
print(df_coll_fin)
```

```python
# Eliminar la columna original 'belongs_to_collection'
df_movies.drop('belongs_to_collection', axis=1)
print(df_movies)
```

```python
# voy a crear una funcion para convertir a diccionario
def convertir_Diccionario(diccionarios):    
    if isinstance(diccionarios, str):
        try:
            diccionarios = ast.literal_eval(diccionarios)
        except (SyntaxError, ValueError):
            return []
    if not isinstance(diccionarios, list):
        return []
    resultado = []
    for diccionario in diccionarios:
        resultado.append(diccionario)
    return resultado
df_credits['cast'] = df_credits['cast'].apply(convertir_Diccionario)
print(df_credits['cast'])
```

```python
# Usamos la funcion creada, para convertir de diccionario a DATAFRAME
df_credits['cast'] = df_credits['cast'].apply(convertir_Diccionario)
df_cast = df_credits['cast'].explode().apply(pd.Series)
print(df_cast)
```

```python
#listamos y vemos las 10 primeras filas
print(df_cast.head(20))
```

```python
df_cast.dtypes
```

```python
df_cast['cast_id'] = df_cast['cast_id'].fillna(0).astype(int)
df_cast['gender'] = df_cast['gender'].fillna(0).astype(int)
df_cast['order'] = df_cast['order'].fillna(0).astype(int)
df_cast['id'] = df_cast['id'].fillna(0).astype(int)
#volvemos a chequear si se cambiaron el tipo de datos
df_cast.dtypes
print(df_cast.head(5))


```

```python
# Unimos en una sola fila los nombre, que tengan igual indice 
df_index= df_cast.groupby(df_cast.index)['name'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
print(df_index)
```

```python
df_index1 = df_index.rename(columns = {'name' : 'actors'})
# Creamos el DATAFRAME para actores
df_actores = df_index1[['actors']]
df_actores[:20]
print(df_index1)
```

```python
#Unimos el nuevo DATAFRAME de actores, con el indice obtenido
df_final_act = df_cast[['id']].join(df_index1, how='left')
df_final_act[:20]
```
